```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 1. Definimos la URL de la prueba
id_prueba = 5902
url = f"https://votaciones.hcdn.gob.ar/votacion/{id_prueba}"

print(f"Iniciando prueba de extracción en: {url}...\n")

# 2. Hacemos la petición a la página
respuesta = requests.get(url)

# Validamos que la página cargó bien
if respuesta.status_code == 200:
    print("✅ Conexión exitosa. Parseando el HTML...")
    
    # 3. Extraer la tabla principal usando Pandas
    # pd.read_html devuelve una lista con todas las tablas que encuentre
    tablas = pd.read_html(respuesta.text)
    
    # Asumimos que la tabla de los votos es la primera que aparece en el código fuente
    df_votacion = tablas[0] 
    
    # Agregamos una columna con el ID por buenas prácticas
    df_votacion['id_votacion'] = id_prueba
    
    # 4. Extraer metadata básica para el dashboard (Título de la pestaña como ejemplo)
    soup = BeautifulSoup(respuesta.text, 'html.parser')
    titulo_pagina = soup.title.text.strip() if soup.title else "Sin título"
    
    # 5. Mostramos los resultados en la consola
    print("\n--- METADATA DE LA SESIÓN ---")
    print(f"Título detectado: {titulo_pagina}")
    print(f"Total de votos/filas extraídas: {len(df_votacion)}")
    
    print("\n--- PRIMERAS 5 FILAS DE LA TABLA ---")
    print(df_votacion.head())
    
    # 6. Exportamos a CSV
    nombre_archivo = f"../data/votacion_prueba_{id_prueba}.csv"
    # utf-8-sig es clave para que los tildes y ñ se vean bien en Excel
    df_votacion.to_csv(nombre_archivo, index=False, encoding='utf-8-sig') 
    
    print(f"\n💾 ¡Éxito! Se generó el archivo '{nombre_archivo}' en la misma carpeta donde ejecutaste el script.")

else:
    print(f"❌ Error al conectar. Código de estado HTTP: {respuesta.status_code}")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 1. Configuración
id_prueba = 5902
url = f"https://votaciones.hcdn.gob.ar/votacion/{id_prueba}"

print(f"Extrayendo datos de la votación {id_prueba}...")

# 2. Petición a la web
respuesta = requests.get(url)

if respuesta.status_code == 200:
    # 3. Parsear el contenido con BeautifulSoup para la metadata
    soup = BeautifulSoup(respuesta.text, 'html.parser')
    
    # --- EXTRACCIÓN DE METADATA ---
    # Buscamos el título (suele estar en un h1 o h2 dentro de la sección de la votación)
    # Según la estructura de la HCDN, el título está en un div con clase específica o un header
    # Usamos selectores más generales para asegurar captura
    try:
        # Intentamos capturar el título del proyecto
        titulo_voto = soup.find('h2').text.strip() if soup.find('h2') else "Título no encontrado"
        
        # Intentamos capturar la fecha (suele estar en un formato 'dd/mm/aaaa' dentro de un párrafo o span)
        # En esta web, la fecha suele estar cerca de un icono de reloj o en un campo de texto pequeño
        # Buscamos el texto que tiene formato de fecha (xx/xx/xxxx)
        div_datos = soup.find('div', {'class': 'col-md-9'}) # Contenedor principal de la metadata
        fecha_voto = "Fecha no encontrada"
        
        if div_datos:
            # Buscamos dentro del texto del contenedor algún patrón de fecha
            import re
            match_fecha = re.search(r'(\d{2}/\d{2}/\d{4})', div_datos.text)
            if match_fecha:
                fecha_voto = match_fecha.group(1)
    except Exception as e:
        print(f"Aviso: No se pudo extraer alguna metadata: {e}")
        titulo_voto, fecha_voto = "Error", "Error"

    # --- EXTRACCIÓN DE LA TABLA ---
    tablas = pd.read_html(respuesta.text)
    df = tablas[0]
    
    # --- AGREGAR NUEVAS COLUMNAS ---
    # Al asignar un valor a una columna que no existe, Pandas la crea para todas las filas
    df['id_votacion'] = id_prueba
    df['titulo_proyecto'] = titulo_voto
    df['fecha_votacion'] = fecha_voto
    
    # 4. Guardar y mostrar
    print(f"\n✅ Datos recuperados:")
    print(f"📅 Fecha: {fecha_voto}")
    print(f"📜 Proyecto: {titulo_voto[:80]}...") # Cortamos el texto si es muy largo
    
    nombre_archivo = f"../data/votacion_{id_prueba}_completa.csv"
    df.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')
    
    print(f"\n💾 Archivo guardado como: {nombre_archivo}")
    print(df.head(3)) # Mostramos solo las primeras 3 filas para verificar las columnas

else:
    print(f"❌ Error {respuesta.status_code}")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 1. Configuración de la prueba
id_prueba = 5902
url = f"https://votaciones.hcdn.gob.ar/votacion/{id_prueba}"

print(f"Extrayendo datos de la votación {id_prueba}...")

respuesta = requests.get(url)

if respuesta.status_code == 200:
    soup = BeautifulSoup(respuesta.text, 'html.parser')
    
    # --- EXTRACCIÓN PRECISA DE METADATA ---
    try:
        # Buscamos exactamente el <h4> con su clase específica
        etiqueta_titulo = soup.find('h4', class_='black-opacity')
        titulo_voto = etiqueta_titulo.text.strip() if etiqueta_titulo else "Título no encontrado"
    except Exception as e:
        titulo_voto = "Error al extraer título"
        
    try:
        # Buscamos exactamente el <h5> con su clase específica
        etiqueta_fecha = soup.find('h5', class_='text-muted m-t-0')
        if etiqueta_fecha:
            # El texto viene como "09/04/2026 - 02:30". 
            # Usamos split() para cortarlo por el guion y quedarnos solo con la primera parte (la fecha).
            fecha_voto = etiqueta_fecha.text.split('-')[0].strip()
        else:
            fecha_voto = "Fecha no encontrada"
    except Exception as e:
        fecha_voto = "Error al extraer fecha"

    # --- EXTRACCIÓN DE LA TABLA ---
    tablas = pd.read_html(respuesta.text)
    df = tablas[0]
    
    # --- AGREGAR NUEVAS COLUMNAS AL DATASET ---
    df['id_votacion'] = id_prueba
    df['titulo_proyecto'] = titulo_voto
    df['fecha_votacion'] = fecha_voto
    
    # 4. Guardar y mostrar resultados
    print("\n✅ ¡Extracción exitosa!")
    print(f"📅 Fecha detectada: {fecha_voto}")
    print(f"📜 Proyecto detectado: {titulo_voto}")
    
    nombre_archivo = f"../data/votacion_{id_prueba}_completa.csv"
    df.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')
    
    print(f"\n💾 Archivo guardado como: {nombre_archivo}")
    
    # Mostramos las primeras 3 filas para comprobar que las columnas se pegaron bien
    print("\n--- Vista previa de los datos ---")
    print(df[['DIPUTADO', 'BLOQUE', 'fecha_votacion', 'titulo_proyecto']].head(3))

else:
    print(f"❌ Error al conectar con la página. Código: {respuesta.status_code}")
```
